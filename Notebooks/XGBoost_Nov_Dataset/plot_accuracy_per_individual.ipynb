```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##Â Rat Model 

```python

data_dict = {'Animal_Data': ['S7068', 'S7069', 'S7070', 'S7071', 'S7072', 'S7083', 'S7088', 'S7091', 'S7092',
                         'S7094', 'S7101', 'S7076', 'S7086'],
           'Accuracy': [0.450738, 0.982799, 0.895515, 0.938765, 0.895185, 0.998352, 0.780216, 0.976515,
                         0.783074, 0.928795, 0.930187, 0.727393, 0.885462],
           'Dataset': ['Training', 'Training', 'Training', 'Training', 'Training', 'Training', 'Training', 
                      'Training', 'Training', 'Training', 'Training', 'Test', 'Test']} 

plot_data = pd.DataFrame(data = data_dict)
```

```python
# Setting the style and font size
sns.set_style("white", rc={'axes.labelsize': 20, 'axes.titlesize': 20})

palette_bar = ['skyblue', 'orangered']

# Creating the plot
fig, axs = plt.subplots(1, 1, figsize=(10, 8), sharex=True, sharey=True)
sns.barplot(data=plot_data, x='Animal_Data', y='Accuracy', hue='Dataset',
            errorbar=('se'), linewidth=4, ax=axs, dodge=0.9, width = 1.0,
           palette = palette_bar)  # Adjust dodge for bar closeness

# Making axes lines thicker
for axis in ['top','bottom','left','right']:
    axs.spines[axis].set_linewidth(2)  # Adjust linewidth for axes thickness

# Increase the size of the axes labels if needed
axs.set_xlabel('Animal ID', fontsize=20)  # Edit the label text and font size as needed
axs.set_ylabel('Accuracy', fontsize=20)

sns.despine()
#plt.show()

os.chdir('/home/melissa/RESULTS/FINAL_MODEL/Rat/Evaluation_Plots/')
plt.savefig('/home/melissa/RESULTS/FINAL_MODEL/Rat/Evaluation_Plots/accuracy_by_animal.png', format = 'png')
plt.savefig('/home/melissa/RESULTS/FINAL_MODEL/Rat/Evaluation_Plots/accuracy_by_animal.svg', format = 'svg')
```
